Lists have many methods. Methods are functions that are associated with a specific object. The append method adds an element to the end of a list. In the example use the append method to create a list of squares.


````python
nums = [0, 1, 2, 3, 4]
squares = []
for x in nums:
    squares.append(x ** 2)
print(squares)
````

We can also use list comprehension to create a list of squares to simplify the above code

````python
nums = [0, 1, 2, 3, 4]
squares = [x ** 2 for x in nums]
print(squares)
````

````python
nums = [0, 1, 2, 3, 4]
even_squares = [x ** 2 for x in nums if x % 2 == 0]
print(even_squares)
````

````python
## Questions
1. What did the previous code do?

Answer: It created a list of squares, but only for the even numbers (0, 2, 4).
````

## Functions
In Python a function is defined using the def keyword:


````python
def my_function():
  print("Hello from a function")
````

To call a function, use the function name followed by parenthesis:

Information can be passed into functions as arguments.

Arguments are specified after the function name, inside the parentheses. You can add as many arguments as you want, just separate them with a comma.

The following example has a function with one argument (fname). When the function is called, we pass along a first name, which is used inside the function to print the full name:

````python
def my_function(fname):
  print(fname + " Refsnes")

my_function("Emil")
my_function("Tobias")
my_function("Linus")
````

By default, a function must be called with the correct number of arguments. Meaning that if your function expects 2 arguments, you have to call the function with 2 arguments, not more, and not less.

You can also send arguments with the key = value syntax. This way the order of the arguments does not matter.

In addition, an argument can have a default value. If we call the function without argument, it uses the default value

````python
def my_function(name = "Jane", country = "Norway"):
  print(name + " is from " + country)

my_function("Marianne", "Sweden")
my_function(country="India")
my_function()
my_function(name="Jensen")
````

To let a function return a value, use the return statement:

````python
def my_function(x):
  return 5 * x

print(my_function(3))
print(my_function(5))
print(my_function(9))
````

We usually need to filter the data in a dataframe. When filtering, Pandas doesn't necessarily duplicate the relevant cases in memory. Instead, it often creates a view, meaning it reuses the same memory location but restricts access to a specific part of it. While this approach is memory efficient, it can lead to warnings and errors when making modifications to the filtered data later. This concept is illustrated using the same dataset in our demonstration.

````python
df_filtered = df[df['population'] > 1200000000]
df_filtered
````

````python
df_filtered['new'] = [1,2]
````

To solve this issues, we need to make an explicit copy with .copy method before we start any modifications:

````python
df_filtered = df[df['population'] > 1200000000].copy()
df_filtered
````

````python
df_filtered['new'] = [1,2]
````

## Time Series in Pandas
Pandas has extensive support for time series data. A time series is a series of data points indexed in time order.
We will first create a set of dates using the date_range function. The function creates a series of dates with an specified frequency. The following code creates a series of 100 dates with a hourly frequency.

````python
#create a time series
dates = pd.date_range('20130101', periods=100, freq='H') #D for days, M for months, H for hours, S for seconds, check documentation for more
dates
````

We will now create a dataframe in which the index is the time series we just created. The values are random numbers from a normal distribution.

````python
#Create a dataframe with a time index, using random values
df = pd.DataFrame(np.random.randn(100,4), index=dates, columns=list('ABCD'))
df
````

The index of this dataframe is a datetime index. As with other indexes, we can access the data by slicing the index. The following code prints the data between 2013-01-02 00:00:00 and 2013-01-03 23:59:59.

````python
#Filter the date
df['20130102':'20130103']
````

Try accessing the date between 2013-01-02 00:06:00 and 2013-01-02 00:05:59.

````python
# This returns an empty result because the start time is after the end time

df['20130102 00:06:00':'20130102 00:05:59']
````

We can extract the year, month, day, hour, minute and second from the index using its properties. The following code prints the year from the index. You can use the same syntax to access other properties (month, day, hour, minute, second).

````python
#creating a column with the year
df['year'] = df.index.year
df
````

The previous code accessed date properties from a DateTimeIndex. However, if the dates were in a column and not an index, we would have to access the properties using the dt accessor. The following code illustrates this:

````python
#creating a column with the whole date
df['date'] = df.index
df
````

The dt accessor is a property of columns that are of type datetime. The following code prints the year from the date column.

````python
#accresing the year from the date column
df['date'].dt.year
````

One of the most useful features of time series is resampling. Resampling is the process of converting a time series from one frequency to another. To resample a time series, we first need to specify the new frequency and then apply a function to each group to obtain the new value. The following code resamples the data to get the mean of each month. We could instead use sum, max, min, median, etc.

````python
#resampling the data to get the mean of each day
df_resampled = df.resample('D').mean() #D for days, M for months, H for hours, S for seconds, check documentation for more
df_resampled
````
